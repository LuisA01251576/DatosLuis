```python
# Se importan las librerias y se manda a llamar la base de datos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("insurance.csv")
df.head()
```

```python
# df.info da informacion general acerca de la base de datos
df.info()
```

```python
# df.describe te da un peque√±o analisis de la base de datos
df.describe()
```

```python
# Ver como esta el peso de la poblacion 
sns.boxplot(df["bmi"]);
```

```python
# Como el precio del seguro depende de la edad y si fuma o no 
plt.figure(figsize=(7,7))
sns.scatterplot(x="age",y="charges",data=df,hue="smoker");
plt.title("A√±o vs Cargos")
plt.figure();
```

```python
# Costo del seguro segun el sexo y la region
plt.figure(figsize=(7,7))
sns.catplot(x="sex",y="charges",hue="region",data=df, kind="bar")
plt.title("Costo en base a la region")
plt.figure();
```
